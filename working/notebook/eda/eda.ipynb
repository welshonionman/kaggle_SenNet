```python
from glob import glob
import tifffile
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_train = "/kaggle/input/blood-vessel-segmentation/train"
dir_raw = "/kaggle/working/dataset/stack_raw"
dir_viz = "/kaggle/working/dataset/stack_viz"
dir_clipped = "/kaggle/working/dataset/stack_clipped"
```

```python
for npy_path in sorted(glob(f"{dir_raw}/*.npy")):
    data = np.load(npy_path)
    print(npy_path.split("/")[-1], data.shape)
```

```python
for npy_path in glob(f"{dir_raw}/*.npy"):
    if "labels" in npy_path:
        continue
    if "voi" in npy_path:
        continue
    npy = np.load(npy_path)
    plt.figure(figsize=(15, 3))
    plt.hist(npy.flatten(), bins=256)
    plt.title(npy_path.split("/")[-1])
```

```python
for npy_path in glob(f"{dir_clipped}/*.npy"):
    if "labels" in npy_path:
        continue
    if "voi" in npy_path:
        continue
    if "voi" in npy_path:
        continue

    npy = np.load(npy_path)
    plt.figure(figsize=(15, 3))
    plt.hist(npy.flatten(), bins=256)
    plt.title(npy_path.split("/")[-1])
```
